Importing most common misspelt spelling along with their correct spells using birkbeck

```python
from birkbeck import birkbeck

birkbeck_dictionary = birkbeck()
birkbeck_dictionary
```

Calculating the Levenshtein Distance using mindist which uses multiprocessing in its backend. You can change the subset of dictionary here and the desired values of k. 
Success at k is also calculated after that. 

```python
from mindist import min_dist_k
from successatk import success_at_k

k_values = [1,5,10]
results = {}
run = {}

for key, value in birkbeck_dictionary.items():
  temp = {}
  run_temp = {}
  # All words 
  # wordnet = dc.all_words(letter="", length=0)

  # All words with same initial and same length
  wordnet = dc.all_words(letter=key[0], length=len(value))
  
  # All wors with same initial
  # wordnet = dc.all_words(letter=key[0], length=0)

  # All words with same length
  # wordnet = dc.all_words(letter="", length=len(value))
  dist_list = min_dist_k(key, wordnet, max(k_values))
  for k in [1,5,10]:
    suc_at_k = success_at_k(value, dist_list, k)
    temp[str(k)] = suc_at_k
    run_temp[str(k)] = 1
    # print("Word: ",key,"\t\tSuccess at ",k,": ",suc_at_k,"\t\tCorrection: ",value)
  # print("K list :",dist_list,"Len :",len(dist_list))
  results[str(key)] = temp
  run[str(key)] = run_temp
```

Calculating the map, ndcg and success for calculated results for every k

```python
import pytrec_eval
import json  

evaluator = pytrec_eval.RelevanceEvaluator(
    results, {'map', 'ndcg', 'success'})

print(json.dumps(evaluator.evaluate(run), indent=1))
eval = evaluator.evaluate(run)
```

Calculating average for all the values of given k

```python
# print(eval.values())
measures = ["map", "ndcg", "success_1", "success_5", "success_10"]

for k in measures:
  print(k, ": \t\t", pytrec_eval.compute_aggregated_measure(k, [query[k] for query in eval.values()]))
```
